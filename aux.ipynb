```python
import json
import pandas as pd

STRUCTURED_COLS = [
    "lotes",
    "resultado",
    "fecha_acuerdo",
    "ofertas_recibidas",
    "ofertas_pymes",
    "adjudicatario_nombre",
    "identificador",
    "adjudicatario_pyme",
    "adjudicatario_ute",
    "importe_total_sin_iva",
    "importe_total_con_iva",
]

def deserialize_from_parquet(df):
    df = df.copy()
    for c in STRUCTURED_COLS:
        if c in df.columns:
            df[c] = df[c].apply(
                lambda x: json.loads(x) if isinstance(x, str) else None
            )
    return df

parquet_path = "/Users/jarenas/Datasets/PLACE/insiders_2526.parquet"

df = pd.read_parquet(parquet_path, engine="pyarrow")
df = deserialize_from_parquet(df)
```

```python
def print_entries_line_by_line(df):
    print("=" * 120)
    for row in df.itertuples(index=False):
        for col, value in zip(df.columns, row):
            print(f"{col}: {value}")
        print("-" * 120)

print_entries_line_by_line(df.sample(n=5))
```

```python
from urllib.parse import urlparse

def extract_domain(url):
    if isinstance(url, str) and url:
        return urlparse(url).netloc.lower()
    return None

def build_whitelist(df):
    domains = set()
    
    for col in ["pliego_admin_URI", "pliego_tecnico_URI"]:
        if col in df.columns:
            for url in df[col].dropna():
                domain = extract_domain(url)
                if domain:
                    domains.add(domain)
    
    return sorted(domains)

whitelist = build_whitelist(df)

print("White list de dominios necesarios:")
for d in whitelist:
    print(d)
```
