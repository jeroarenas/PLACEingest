```python
import zipfile
from pathlib import Path
import xml.etree.ElementTree as ET
import pandas as pd

def localname(tag: str) -> str:
    return tag.split("}", 1)[1] if tag.startswith("{") else tag

def flatten_xml(elem, prefix="", out=None, join="_"):
    """
    Aplana un árbol XML a un dict.
    - Usa localname(tag) para ignorar namespaces.
    - Incluye atributos como <tag>@attr
    - Si hay repetidos, los indexa: Field, Field_1, Field_2...
    """
    if out is None:
        out = {}

    tag = localname(elem.tag)
    base = f"{prefix}{join}{tag}" if prefix else tag

    # atributos
    for k, v in elem.attrib.items():
        out[f"{base}{join}@{k}"] = v

    children = list(elem)
    text = (elem.text or "").strip()

    if not children:
        # hoja
        if text != "":
            key = base
            if key in out:
                # colisión: indexamos
                i = 1
                while f"{key}{join}{i}" in out:
                    i += 1
                out[f"{key}{join}{i}"] = text
            else:
                out[key] = text
        return out

    # nodo con hijos: si tiene texto "real" también lo guardamos
    if text != "":
        out[f"{base}{join}#text"] = text

    # procesar hijos; si hay repetidos, indexar
    counts = {}
    for ch in children:
        ch_tag = localname(ch.tag)
        counts[ch_tag] = counts.get(ch_tag, 0) + 1

    seen = {}
    for ch in children:
        ch_tag = localname(ch.tag)
        seen[ch_tag] = seen.get(ch_tag, 0) + 1
        # si está repetido, sufijo _{idx}
        ch_prefix = base
        if counts[ch_tag] > 1:
            ch_prefix = f"{base}{join}{ch_tag}{join}{seen[ch_tag]}"
            # y seguimos aplanando bajo ese prefijo "indexado"
            flatten_xml(ch, prefix=ch_prefix, out=out, join=join)
        else:
            flatten_xml(ch, prefix=base, out=out, join=join)

    return out

def parse_atom_from_zip(zip_path: Path, atom_name: str):
    """
    Devuelve lista de dicts (1 dict por entry).
    Incluye campos Atom + flatten de ContractFolderStatus.
    """
    records = []
    with zipfile.ZipFile(zip_path, "r") as z:
        with z.open(atom_name) as f:
            tree = ET.parse(f)
            root = tree.getroot()

    for entry in root:
        if localname(entry.tag) != "entry":
            continue

        rec = {
            "id": None,
            "link": None,
            "summary": None,
            "title": None,
            "updated": None,
            "_source_zip": zip_path.name,
            "_source_atom": atom_name,
        }

        # hijos directos del entry
        for ch in list(entry):
            ln = localname(ch.tag)

            if ln == "id":
                rec["id"] = (ch.text or "").strip()

            elif ln == "updated":
                rec["updated"] = (ch.text or "").strip()

            elif ln == "title":
                rec["title"] = (ch.text or "").strip()

            elif ln == "summary":
                rec["summary"] = (ch.text or "").strip()

            elif ln == "link":
                rec["link"] = ch.attrib.get("href")

            elif ln == "ContractFolderStatus":
                flat = flatten_xml(ch, prefix="ContractFolderStatus")
                rec.update(flat)

        records.append(rec)

    return records
```

```python
from pathlib import Path
import zipfile
import xml.etree.ElementTree as ET
import pandas as pd
import time
from concurrent.futures import ProcessPoolExecutor, as_completed
import os

BASE = Path("/home/jarenas/Datasets/PLACE/2025")
zip_paths = sorted(BASE.glob("licitacionesPerfilesContratanteCompleto3_2025*.zip"))

def localname(tag: str) -> str:
    return tag.split("}", 1)[1] if tag.startswith("{") else tag

def list_atom_files_in_zip(zip_path: Path):
    with zipfile.ZipFile(zip_path, "r") as z:
        return [n for n in z.namelist() if n.lower().endswith(".atom")]

def process_one_atom(task):
    """
    task: (zip_path_str, atom_name, collect_ids)
    returns: (zip_name, atom_name, n_entries, ids_list_or_None)
    """
    zip_path_str, atom_name, collect_ids = task
    zip_path = Path(zip_path_str)

    with zipfile.ZipFile(zip_path, "r") as z:
        with z.open(atom_name) as f:
            root = ET.parse(f).getroot()

    n_entries = 0
    ids = [] if collect_ids else None

    for entry in list(root):
        if localname(entry.tag) != "entry":
            continue
        n_entries += 1
        if collect_ids:
            for ch in list(entry):
                if localname(ch.tag) == "id":
                    ids.append((ch.text or "").strip())
                    break

    return zip_path.name, atom_name, n_entries, ids

# -------- CONFIG --------
COLLECT_IDS = False                 # True si quieres sets de ids
MAX_WORKERS = min(20, os.cpu_count() or 1)
PROGRESS_EVERY = 25                 # imprime cada 25 atoms completados
# ------------------------

# Build tasks
tasks = []
zip_to_atoms = {}
for zp in zip_paths:
    atoms = list_atom_files_in_zip(zp)
    zip_to_atoms[zp.name] = len(atoms)
    for a in atoms:
        tasks.append((str(zp), a, COLLECT_IDS))

print("ZIPs:", len(zip_paths))
print("Total atom files:", len(tasks))
print("Workers:", MAX_WORKERS)

t0 = time.time()

# Aggregators
zip_entries = {zp.name: 0 for zp in zip_paths}
zip_atoms_done = {zp.name: 0 for zp in zip_paths}
total_entries = 0

all_ids = [] if COLLECT_IDS else None

done = 0
with ProcessPoolExecutor(max_workers=MAX_WORKERS) as ex:
    futures = [ex.submit(process_one_atom, t) for t in tasks]

    for fut in as_completed(futures):
        zip_name, atom_name, n_entries, ids = fut.result()

        done += 1
        total_entries += n_entries
        zip_entries[zip_name] += n_entries
        zip_atoms_done[zip_name] += 1

        if COLLECT_IDS and ids:
            all_ids.extend(ids)

        if done % PROGRESS_EVERY == 0 or done == len(tasks):
            elapsed = time.time() - t0
            # muestra también el ZIP "más avanzado" por completitud
            # (opcional: puedes comentar esto si molesta)
            z_sample = max(zip_atoms_done, key=lambda k: zip_atoms_done[k])
            print(
                f"[{done}/{len(tasks)} atoms] total_entries={total_entries:,} "
                f"| elapsed={elapsed:,.1f}s | sample_zip={z_sample} "
                f"({zip_atoms_done[z_sample]}/{zip_to_atoms[z_sample]} atoms)"
            )

elapsed = time.time() - t0
print("\nDONE. Total entries:", total_entries, "| elapsed:", f"{elapsed:.1f}s")

# Summary per ZIP
df_zip_counts = pd.DataFrame(
    [(z, zip_to_atoms[z], zip_entries[z], zip_atoms_done[z]) for z in zip_entries.keys()],
    columns=["zip", "n_atom_files", "n_entries", "atoms_done"]
).sort_values("zip")

print("\n=== RESUMEN POR ZIP ===")
print(df_zip_counts)

# Compare with parquet counts
df_insiders = pd.read_parquet(BASE / "insiders.parquet", engine="fastparquet")
df_unif_old = pd.read_parquet(BASE / "insiders_unificado.parquet", engine="fastparquet")

print("\nParquet rows:")
print("insiders:", df_insiders.shape[0], "unique ids:", df_insiders["id"].nunique())
print("unif_old:", df_unif_old.shape[0], "unique ids:", df_unif_old["id"].nunique())

if COLLECT_IDS:
    s_atom = pd.Series(all_ids, name="id")
    print("\nAtom ids:")
    print("n ids:", len(s_atom), "unique ids:", s_atom.nunique(), "dups:", s_atom.duplicated().sum())

    ids_atom = set(s_atom.unique())
    ids_ins = set(df_insiders["id"].astype(str).unique())
    ids_unf = set(df_unif_old["id"].astype(str).unique())

    print("\nID set comparison:")
    print("Atom - insiders:", len(ids_atom - ids_ins))
    print("Insiders - Atom:", len(ids_ins - ids_atom))
    print("Atom - unif_old:", len(ids_atom - ids_unf))
    print("Unif_old - Atom:", len(ids_unf - ids_atom))
```

```python
from pathlib import Path
import zipfile
import xml.etree.ElementTree as ET
import pandas as pd
import time
from concurrent.futures import ProcessPoolExecutor, as_completed
import os
from collections import Counter

BASE = Path("/home/jarenas/Datasets/PLACE/2025")
zip_paths = sorted(BASE.glob("licitacionesPerfilesContratanteCompleto3_2025*.zip"))

def localname(tag: str) -> str:
    return tag.split("}", 1)[1] if tag.startswith("{") else tag

def list_atom_files_in_zip(zip_path: Path):
    with zipfile.ZipFile(zip_path, "r") as z:
        return [n for n in z.namelist() if n.lower().endswith(".atom")]

def process_atom_ids(task):
    zip_path_str, atom_name = task
    zip_path = Path(zip_path_str)

    with zipfile.ZipFile(zip_path, "r") as z:
        with z.open(atom_name) as f:
            root = ET.parse(f).getroot()

    ids = []
    for entry in list(root):
        if localname(entry.tag) != "entry":
            continue
        for ch in list(entry):
            if localname(ch.tag) == "id":
                ids.append((ch.text or "").strip())
                break
    return zip_path.name, atom_name, ids

# tareas
tasks = []
for zp in zip_paths:
    for a in list_atom_files_in_zip(zp):
        tasks.append((str(zp), a))

MAX_WORKERS = min(20, os.cpu_count() or 1)
PROGRESS_EVERY = 25

t0 = time.time()
counter = Counter()
done = 0

with ProcessPoolExecutor(max_workers=MAX_WORKERS) as ex:
    futures = [ex.submit(process_atom_ids, t) for t in tasks]
    for fut in as_completed(futures):
        _, _, ids = fut.result()
        counter.update(ids)
        done += 1
        if done % PROGRESS_EVERY == 0 or done == len(tasks):
            print(f"[{done}/{len(tasks)}] unique_ids={len(counter):,} elapsed={time.time()-t0:.1f}s")

total_entries = sum(counter.values())
unique_ids = len(counter)
dup_entries = total_entries - unique_ids
print("\nAtom totals:")
print("entries:", total_entries)
print("unique ids:", unique_ids)
print("duplicate entries:", dup_entries)
print("avg entries per id:", total_entries / unique_ids)

# comparar con parquet
df_insiders = pd.read_parquet(BASE / "insiders.parquet", engine="fastparquet")
ids_ins = set(df_insiders["id"].astype(str).unique())

ids_atom = set(counter.keys())
print("\nSet comparison with insiders.parquet:")
print("Atom - insiders:", len(ids_atom - ids_ins))
print("Insiders - Atom:", len(ids_ins - ids_atom))
print("Examples Atom - insiders:", list(sorted(ids_atom - ids_ins))[:5])
print("Examples Insiders - Atom:", list(sorted(ids_ins - ids_atom))[:5])

# opcional: ver los top IDs más repetidos en Atom (para confirmar naturaleza de duplicado)
print("\nTop repeated Atom ids:")
print(counter.most_common(10))
```

```python
from pathlib import Path
import zipfile
import xml.etree.ElementTree as ET
import pandas as pd
import os
from concurrent.futures import ProcessPoolExecutor, as_completed
from tqdm.auto import tqdm

BASE = Path("/home/jarenas/Datasets/PLACE/2025")
zip_paths = sorted(BASE.glob("licitacionesPerfilesContratanteCompleto3_2025*.zip"))

def localname(tag: str) -> str:
    return tag.split("}", 1)[1] if tag.startswith("{") else tag

def list_atom_files_in_zip(zip_path: Path):
    with zipfile.ZipFile(zip_path, "r") as z:
        return [n for n in z.namelist() if n.lower().endswith(".atom")]

def parse_atom_light(task):
    """
    task: (zip_path_str, atom_name)
    return: list[dict] (1 dict por entry)
    """
    zip_path_str, atom_name = task
    zip_path = Path(zip_path_str)

    with zipfile.ZipFile(zip_path, "r") as z:
        with z.open(atom_name) as f:
            root = ET.parse(f).getroot()

    records = []
    for entry in list(root):
        if localname(entry.tag) != "entry":
            continue

        rec = {
            "id": None,
            "link": None,
            "summary": None,
            "title": None,
            "updated": None,
            "_source_zip": zip_path.name,
            "_source_atom": atom_name,
            "_contract_folder_status_xml": None,
        }

        for ch in list(entry):
            ln = localname(ch.tag)
            if ln == "id":
                rec["id"] = (ch.text or "").strip()
            elif ln == "updated":
                rec["updated"] = (ch.text or "").strip()
            elif ln == "title":
                rec["title"] = (ch.text or "").strip()
            elif ln == "summary":
                rec["summary"] = (ch.text or "").strip()
            elif ln == "link":
                rec["link"] = ch.attrib.get("href")
            elif ln == "ContractFolderStatus":
                rec["_contract_folder_status_xml"] = ET.tostring(ch, encoding="unicode")

        records.append(rec)

    return records

# tareas (zip, atom)
tasks = []
for zp in zip_paths:
    for atom in list_atom_files_in_zip(zp):
        tasks.append((str(zp), atom))

MAX_WORKERS = min(20, os.cpu_count() or 1)

all_records = []
with ProcessPoolExecutor(max_workers=MAX_WORKERS) as ex:
    futures = [ex.submit(parse_atom_light, t) for t in tasks]
    for fut in tqdm(as_completed(futures), total=len(futures), desc="Parsing .atom"):
        all_records.extend(fut.result())

df_atom_light = pd.DataFrame(all_records)
print("df_atom_light:", df_atom_light.shape)
```

## 2.1. Función para leer los datos y consolidar con lo que ya hubiera

### 2.1.1. Versión monocore

```python
from __future__ import annotations

from pathlib import Path
from typing import Iterable, Union, Optional, Dict, Any, List
import zipfile
import xml.etree.ElementTree as ET
import pandas as pd
from tqdm.auto import tqdm


def _localname(tag: str) -> str:
    return tag.split("}", 1)[1] if tag.startswith("{") else tag


def _iter_input_paths(sources: Union[str, Path, Iterable[Union[str, Path]]]) -> List[Path]:
    """
    Normaliza 'sources' a una lista de Paths:
    - acepta Path/str, lista de Path/str
    - acepta directorios (recursivo: *.zip, *.atom)
    - acepta globs en str (p.ej. '/ruta/*.zip')
    """
    if isinstance(sources, (str, Path)):
        sources = [sources]

    out: List[Path] = []
    for s in sources:
        p = Path(s)
        if isinstance(s, str) and any(ch in s for ch in ["*", "?", "[", "]"]):
            out.extend([Path(x) for x in sorted(Path().glob(s))])
            continue

        if p.is_dir():
            out.extend(sorted(p.rglob("*.zip")))
            out.extend(sorted(p.rglob("*.atom")))
        else:
            out.append(p)

    # de-dup preservando orden
    seen = set()
    deduped = []
    for p in out:
        rp = p.resolve()
        if rp not in seen:
            seen.add(rp)
            deduped.append(rp)
    return deduped


def _parse_atom_fileobj_to_latest_map(fileobj, latest_by_id: Dict[str, Dict[str, Any]]) -> None:
    """
    Lee un ATOM (XML) desde un file-like y actualiza latest_by_id[id] con el registro más reciente por updated.
    Extrae: id, title, summary, updated, link
    """
    root = ET.parse(fileobj).getroot()

    for entry in list(root):
        if _localname(entry.tag) != "entry":
            continue

        rec = {"id": None, "title": None, "summary": None, "updated": None, "link": None}

        for ch in list(entry):
            ln = _localname(ch.tag)
            if ln == "id":
                rec["id"] = (ch.text or "").strip()
            elif ln == "title":
                rec["title"] = (ch.text or "").strip()
            elif ln == "summary":
                rec["summary"] = (ch.text or "").strip()
            elif ln == "updated":
                rec["updated"] = (ch.text or "").strip()
            elif ln == "link":
                # en estos feeds suele haber un único link con href
                if rec["link"] is None:
                    rec["link"] = ch.attrib.get("href")

        if not rec["id"] or not rec["updated"]:
            continue

        upd = pd.to_datetime(rec["updated"], errors="coerce", utc=True)
        if pd.isna(upd):
            continue

        prev = latest_by_id.get(rec["id"])
        if prev is None or upd > prev["_updated_dt"]:
            rec["_updated_dt"] = upd
            latest_by_id[rec["id"]] = rec


def _iter_atom_jobs(paths: list[Path]):
    """
    Genera trabajos:
    - ("zip", zip_path, atom_name)
    - ("atom", atom_path, None)
    """
    for p in paths:
        if p.suffix.lower() == ".zip":
            with zipfile.ZipFile(p, "r") as z:
                atom_names = [n for n in z.namelist() if n.lower().endswith(".atom")]
            for atom_name in atom_names:
                yield ("zip", p, atom_name)
        elif p.suffix.lower() == ".atom":
            yield ("atom", p, None)


def _ingest_sources_to_latest_map(paths: List[Path], show_progress: bool = True) -> Dict[str, Dict[str, Any]]:
    """
    Procesa paths (zip/atom) y devuelve mapa id->registro_latest.
    """
    latest_by_id: Dict[str, Dict[str, Any]] = {}

    jobs = list(_iter_atom_jobs(paths))
    it = tqdm(jobs, desc="Parsing .atom", unit="atom") if show_progress else jobs

    # Cache de zipfile abiertos por ZIP para no reabrirlos cada atom (acelera bastante)
    zcache: Dict[Path, zipfile.ZipFile] = {}
    try:
        for kind, p, atom_name in it:
            if kind == "zip":
                zf = zcache.get(p)
                if zf is None:
                    zf = zipfile.ZipFile(p, "r")
                    zcache[p] = zf
                with zf.open(atom_name) as f:
                    _parse_atom_fileobj_to_latest_map(f, latest_by_id)
            else:
                with p.open("rb") as f:
                    _parse_atom_fileobj_to_latest_map(f, latest_by_id)
    finally:
        for zf in zcache.values():
            zf.close()

    return latest_by_id
    

def _latest_map_to_df(latest_by_id: Dict[str, Dict[str, Any]]) -> pd.DataFrame:
    df = pd.DataFrame(latest_by_id.values())
    if df.empty:
        return df
    # limpiar columna interna
    df = df.drop(columns=["_updated_dt"], errors="ignore")
    # asegurar tipos
    df["id"] = df["id"].astype(str)
    # orden opcional (por updated desc)
    df["_updated_dt"] = pd.to_datetime(df["updated"], errors="coerce", utc=True)
    df = df.sort_values("_updated_dt", ascending=False).drop(columns=["_updated_dt"])
    df = df.reset_index(drop=True)
    return df


def build_latest_entries(
    sources: Union[str, Path, Iterable[Union[str, Path]]],
    parquet_path: Optional[Union[str, Path]] = None,
    *,
    show_progress: bool = True,
) -> pd.DataFrame:
    """
    - sources: fichero(s) zip/atom, carpeta, glob, o lista mezclada.
    - parquet_path (opcional): si se pasa, consolida con lo existente y guarda resultado.
      Criterio siempre: conservar por 'id' la fila con updated más reciente.
    - Devuelve el DataFrame consolidado.
    """
    paths = _iter_input_paths(sources)

    # 1) ingest de sources -> latest map
    latest_map = _ingest_sources_to_latest_map(paths, show_progress=show_progress)

    # 2) si hay parquet, consolidar
    if parquet_path is not None:
        parquet_path = Path(parquet_path)

        if parquet_path.exists():
            df_old = pd.read_parquet(parquet_path, engine="fastparquet")
            # consolidar df_old dentro del mismo mapa (sin reparsear XML)
            # usamos el mismo criterio de updated
            df_old = df_old.copy()
            df_old["_updated_dt"] = pd.to_datetime(df_old["updated"], errors="coerce", utc=True)

            for r in df_old.itertuples(index=False):
                rid = getattr(r, "id", None)
                upd = getattr(r, "_updated_dt", None)
                if rid is None or pd.isna(upd):
                    continue

                rec = {
                    "id": str(rid),
                    "title": getattr(r, "title", None),
                    "summary": getattr(r, "summary", None),
                    "updated": getattr(r, "updated", None),
                    "link": getattr(r, "link", None),
                    "_updated_dt": upd,
                }

                prev = latest_map.get(rec["id"])
                if prev is None or rec["_updated_dt"] > prev["_updated_dt"]:
                    latest_map[rec["id"]] = rec

        # 3) construir df final y guardar
        df_final = _latest_map_to_df(latest_map)
        parquet_path.parent.mkdir(parents=True, exist_ok=True)
        df_final.to_parquet(parquet_path, engine="fastparquet", index=False)
        return df_final

    # sin parquet: solo dataframe
    return _latest_map_to_df(latest_map)
```

#### 2.1.2.2. Helpers XML + normalización

```python
import xml.etree.ElementTree as ET
from datetime import datetime
from urllib.parse import urlparse
from pathlib import Path
import zipfile
import pandas as pd
from concurrent.futures import ProcessPoolExecutor, as_completed
from tqdm import tqdm

def localname(tag: str) -> str:
    return tag.split("}", 1)[1] if tag.startswith("{") else tag

def find_child_by_localname(elem, name):
    for ch in list(elem):
        if localname(ch.tag) == name:
            return ch
    return None

def find_first_text_from(elem, path):
    cur = elem
    for name in path:
        cur = find_child_by_localname(cur, name)
        if cur is None:
            return None
    return (cur.text or "").strip()

def find_all_texts_from(elem, path):
    nodes = [elem]
    for name in path:
        nxt = []
        for n in nodes:
            for ch in list(n):
                if localname(ch.tag) == name:
                    nxt.append(ch)
        nodes = nxt
    return [(n.text or "").strip() for n in nodes if (n.text or "").strip()]

# Normalización de los campos de presupuestos
def to_float(x):
    try:
        return float(x)
    except Exception:
        return None

# Normalización del campo "Plazo de ejecución"
def duration_to_days(start, end, duration, unit):
    # prioridad: si hay start+end -> diff días
    if start and end:
        try:
            d0 = datetime.fromisoformat(start)
            d1 = datetime.fromisoformat(end)
            return (d1 - d0).days
        except Exception:
            pass

    # si hay duration + unit
    if duration and unit:
        try:
            v = float(duration)
            unit = unit.upper()
            if unit == "DAY":
                return int(v)
            if unit == "MON":
                return int(v * 30)
            if unit == "ANN":
                return int(v * 365)
        except Exception:
            pass

    return None
```

#### 2.1.2.3. Extractor de campos

```python
def extract_normalized_fields(entry):
    cfs = find_child_by_localname(entry, "ContractFolderStatus")
    if cfs is None:
        return None

    # campos base
    estado = find_first_text_from(cfs, ["ContractFolderStatusCode"])
    expediente = find_first_text_from(cfs, ["ContractFolderID"])
    objeto = find_first_text_from(cfs, ["ProcurementProject", "Name"])

    valor_estimado = to_float(find_first_text_from(cfs, ["ProcurementProject", "BudgetAmount", "EstimatedOverallContractAmount"]))
    presupuesto_sin_iva = to_float(find_first_text_from(cfs, ["ProcurementProject", "BudgetAmount", "TaxExclusiveAmount"]))
    presupuesto_con_iva = to_float(find_first_text_from(cfs, ["ProcurementProject", "BudgetAmount", "TotalAmount"]))

    # plazo -> días
    start = find_first_text_from(cfs, ["ProcurementProject", "PlannedPeriod", "StartDate"])
    end = find_first_text_from(cfs, ["ProcurementProject", "PlannedPeriod", "EndDate"])
    duration = find_first_text_from(cfs, ["ProcurementProject", "PlannedPeriod", "DurationMeasure"])
    unit = None
    pp = find_child_by_localname(find_child_by_localname(cfs, "ProcurementProject"), "PlannedPeriod")
    if pp is not None:
        dm = find_child_by_localname(pp, "DurationMeasure")
        if dm is not None:
            unit = dm.attrib.get("unitCode")

    duracion_dias = duration_to_days(start, end, duration, unit)

    # CPV
    cpv_list = find_all_texts_from(cfs, ["ProcurementProject", "RequiredCommodityClassification", "ItemClassificationCode"])

    # pliegos
    pliego_tecnico_uri = find_first_text_from(cfs, ["LegalDocumentReference", "Attachment", "ExternalReference", "URI"])
    pliego_tecnico_filename = find_first_text_from(cfs, ["LegalDocumentReference", "ID"])
    
    pliego_admin_uri = find_first_text_from(cfs, ["TechnicalDocumentReference", "Attachment", "ExternalReference", "URI"])
    pliego_admin_filename = find_first_text_from(cfs, ["TechnicalDocumentReference", "ID"])

    # TED UUID
    ted_uuid = None
    for ch in list(cfs):
        if localname(ch.tag) == "UUID" and ch.attrib.get("schemeName") == "TED":
            ted_uuid = (ch.text or "").strip()

    return {
        "estado": estado,
        "expediente": expediente,
        "objeto": objeto,
        "valor_estimado": valor_estimado,
        "presupuesto_sin_iva": presupuesto_sin_iva,
        "presupuesto_con_iva": presupuesto_con_iva,
        "duracion_dias": duracion_dias,
        "cpv_list": cpv_list,
        "pliego_tecnico_uri": pliego_tecnico_uri,
        "pliego_tecnico_filename": pliego_tecnico_filename,
        "pliego_admin_uri": pliego_admin_uri,
        "pliego_admin_filename": pliego_admin_filename,
        "ted_uuid": ted_uuid,
    }
```

#### 2.1.2.4. Extracción multicore de todo

```python
def extract_atom_fields(entry):
    atom = {"title": None, "summary": None, "link": None}
    for ch in list(entry):
        name = localname(ch.tag)
        if name == "title":
            atom["title"] = (ch.text or "").strip()
        elif name == "summary":
            atom["summary"] = (ch.text or "").strip()
        elif name == "link":
            atom["link"] = ch.attrib.get("href")
    return atom
    

def process_zip_for_latest(zip_path):
    latest = {}
    with zipfile.ZipFile(zip_path, "r") as z:
        atoms = [n for n in z.namelist() if n.lower().endswith(".atom")]
        for atom in atoms:
            with z.open(atom) as f:
                root = ET.parse(f).getroot()

            for entry in list(root):
                eid = None
                updated = None
                for ch in list(entry):
                    if localname(ch.tag) == "id":
                        eid = (ch.text or "").strip()
                    elif localname(ch.tag) == "updated":
                        updated = (ch.text or "").strip()

                if not eid or not updated:
                    continue

                atom_fields = extract_atom_fields(entry)
                rec = extract_normalized_fields(entry)
                if rec is None:
                    continue
                
                latest[eid] = {
                    "id": eid,
                    "updated": updated,
                    **atom_fields,
                    **rec,
                    "_updated": updated
                }

    return latest
```

```python
def build_latest_entries_multicore(sources, parquet_path=None, max_workers=8, show_progress=True):
    paths = [Path(p) for p in sources]

    merged = {}

    with ProcessPoolExecutor(max_workers=max_workers) as ex:
        futures = [ex.submit(process_zip_for_latest, p) for p in paths]
        it = as_completed(futures)
        if show_progress:
            it = tqdm(it, total=len(futures), desc="Zips")

        for fut in it:
            latest_part = fut.result()
            for eid, rec in latest_part.items():
                prev = merged.get(eid)
                if (prev is None) or (rec["_updated"] > prev["_updated"]):
                    merged[eid] = rec

    df = pd.DataFrame([v for v in merged.values()]).drop(columns=["_updated"])

    # renombrar columnas según COLUMN_MAP
    df = df.rename(columns=COLUMN_MAP)

    if parquet_path:
        parquet_path = Path(parquet_path)
        if parquet_path.exists():
            old = pd.read_parquet(parquet_path, engine="fastparquet")
            df = (
                pd.concat([old, df], ignore_index=True)
                .sort_values("updated")
                .drop_duplicates(subset=["id"], keep="last")
            )
        df.to_parquet(parquet_path, index=False, engine="fastparquet")

    return df
```

#### 2.1.2.X [DEPRECATED] Código multicore para la extracción únicamente de los datos generales del entry

```python
from __future__ import annotations

from pathlib import Path
from typing import Iterable, Union, Optional, Dict, Any, List, Tuple
import zipfile
import xml.etree.ElementTree as ET
import pandas as pd
import os
from concurrent.futures import ProcessPoolExecutor, as_completed
from tqdm.auto import tqdm
from glob import glob


def _localname(tag: str) -> str:
    return tag.split("}", 1)[1] if tag.startswith("{") else tag


def _iter_input_paths(sources):
    if isinstance(sources, (str, Path)):
        sources = [sources]

    out = []
    for s in sources:
        if isinstance(s, str) and any(ch in s for ch in ["*", "?", "[", "]"]):
            out.extend([Path(x) for x in sorted(glob(s))])
            continue

        p = Path(s)
        if p.is_dir():
            out.extend(sorted(p.rglob("*.zip")))
            out.extend(sorted(p.rglob("*.atom")))
        else:
            out.append(p)

    seen = set()
    deduped = []
    for p in out:
        rp = p.resolve()
        if rp not in seen:
            seen.add(rp)
            deduped.append(rp)
    return deduped


def _iter_atom_jobs(paths: List[Path]) -> List[Tuple[str, str, Optional[str]]]:
    """
    Devuelve lista de jobs:
      - ("zip", "<zip_path>", "<atom_name_inside_zip>")
      - ("atom", "<atom_file_path>", None)
    """
    jobs: List[Tuple[str, str, Optional[str]]] = []
    for p in paths:
        if p.suffix.lower() == ".zip":
            with zipfile.ZipFile(p, "r") as z:
                atom_names = [n for n in z.namelist() if n.lower().endswith(".atom")]
            for atom_name in atom_names:
                jobs.append(("zip", str(p), atom_name))
        elif p.suffix.lower() == ".atom":
            jobs.append(("atom", str(p), None))
    return jobs


def _parse_one_atom_job(job: Tuple[str, str, Optional[str]]) -> Dict[str, Dict[str, Any]]:
    """
    Procesa UN .atom (suelo o dentro de zip) y devuelve un dict:
      id -> {"id","title","summary","updated","link","_updated_dt"}
    Mantiene solo el latest por id dentro de ese atom.
    """
    kind, path_str, atom_name = job

    if kind == "zip":
        with zipfile.ZipFile(path_str, "r") as z:
            with z.open(atom_name) as f:
                root = ET.parse(f).getroot()
    else:
        with open(path_str, "rb") as f:
            root = ET.parse(f).getroot()

    latest: Dict[str, Dict[str, Any]] = {}

    for entry in list(root):
        if _localname(entry.tag) != "entry":
            continue

        rec = {"id": None, "title": None, "summary": None, "updated": None, "link": None}

        for ch in list(entry):
            ln = _localname(ch.tag)
            if ln == "id":
                rec["id"] = (ch.text or "").strip()
            elif ln == "title":
                rec["title"] = (ch.text or "").strip()
            elif ln == "summary":
                rec["summary"] = (ch.text or "").strip()
            elif ln == "updated":
                rec["updated"] = (ch.text or "").strip()
            elif ln == "link":
                if rec["link"] is None:
                    rec["link"] = ch.attrib.get("href")

        if not rec["id"] or not rec["updated"]:
            continue

        upd = pd.to_datetime(rec["updated"], errors="coerce", utc=True)
        if pd.isna(upd):
            continue

        prev = latest.get(rec["id"])
        if prev is None or upd > prev["_updated_dt"]:
            rec["_updated_dt"] = upd
            latest[rec["id"]] = rec

    return latest


def _merge_latest_maps(dst: Dict[str, Dict[str, Any]], src: Dict[str, Dict[str, Any]]) -> None:
    for k, rec in src.items():
        prev = dst.get(k)
        if prev is None or rec["_updated_dt"] > prev["_updated_dt"]:
            dst[k] = rec


def _latest_map_to_df(latest_by_id: Dict[str, Dict[str, Any]]) -> pd.DataFrame:
    df = pd.DataFrame(latest_by_id.values())
    if df.empty:
        return df
    df = df.drop(columns=["_updated_dt"], errors="ignore")
    df["id"] = df["id"].astype(str)
    # orden opcional por updated desc
    tmp = pd.to_datetime(df["updated"], errors="coerce", utc=True)
    df = df.assign(_tmp=tmp).sort_values("_tmp", ascending=False).drop(columns=["_tmp"])
    return df.reset_index(drop=True)


def _ingest_sources_to_latest_map_multicore(
    paths: List[Path],
    *,
    show_progress: bool = True,
    max_workers: Optional[int] = None,
) -> Dict[str, Dict[str, Any]]:
    jobs = _iter_atom_jobs(paths)
    if not jobs:
        return {}

    if max_workers is None:
        max_workers = min(20, os.cpu_count() or 1)

    latest_by_id: Dict[str, Dict[str, Any]] = {}

    with ProcessPoolExecutor(max_workers=max_workers) as ex:
        futures = [ex.submit(_parse_one_atom_job, job) for job in jobs]
        it = as_completed(futures)
        if show_progress:
            it = tqdm(it, total=len(futures), desc="Parsing .atom", unit="atom")

        for fut in it:
            local_map = fut.result()
            _merge_latest_maps(latest_by_id, local_map)

    return latest_by_id


def build_latest_entries(
    sources: Union[str, Path, Iterable[Union[str, Path]]],
    parquet_path: Optional[Union[str, Path]] = None,
    *,
    show_progress: bool = True,
    max_workers: Optional[int] = None,
) -> pd.DataFrame:
    """
    Multicore por .atom. Conserva la entrada más reciente por id (updated max).
    Si parquet_path:
      - si no existe, crea
      - si existe, lee y consolida (latest por id)
    Devuelve el DF consolidado.
    """
    paths = _iter_input_paths(sources)

    # 1) ingest multicore
    latest_map = _ingest_sources_to_latest_map_multicore(
        paths, show_progress=show_progress, max_workers=max_workers
    )

    # 2) consolidación con parquet (si aplica)
    if parquet_path is not None:
        parquet_path = Path(parquet_path)

        if parquet_path.exists():
            df_old = pd.read_parquet(parquet_path, engine="fastparquet").copy()
            df_old["_updated_dt"] = pd.to_datetime(df_old["updated"], errors="coerce", utc=True)

            for r in df_old.itertuples(index=False):
                rid = getattr(r, "id", None)
                upd = getattr(r, "_updated_dt", None)
                if rid is None or pd.isna(upd):
                    continue

                rec = {
                    "id": str(rid),
                    "title": getattr(r, "title", None),
                    "summary": getattr(r, "summary", None),
                    "updated": getattr(r, "updated", None),
                    "link": getattr(r, "link", None),
                    "_updated_dt": upd,
                }

                prev = latest_map.get(rec["id"])
                if prev is None or rec["_updated_dt"] > prev["_updated_dt"]:
                    latest_map[rec["id"]] = rec

        df_final = _latest_map_to_df(latest_map)
        parquet_path.parent.mkdir(parents=True, exist_ok=True)
        df_final.to_parquet(parquet_path, engine="fastparquet", index=False)
        return df_final

    return _latest_map_to_df(latest_map)
```

```python
import zipfile
import xml.etree.ElementTree as ET
import pandas as pd
from pathlib import Path

# ---------- Config ----------
BASE = Path("/home/jarenas/Datasets/PLACE/2025")
zip_path = sorted(BASE.glob("licitacionesPerfilesContratanteCompleto3_2025*.zip"))[0]
MAX_ENTRIES = 50  # ajusta: 10, 50, 200...

PROCEDURE_CODE_MAP = {
    "1": "Abierto",
    "2": "Restringido",
    "3": "Negociado sin publicidad",
    "4": "Negociado con publicidad",
    "5": "Diálogo competitivo",
    "6": "Contrato menor",
    "7": "Derivado de acuerdo marco",
    "8": "Concurso de proyectos",
    "100": "Normas internas",
    "999": "Otros",
}
URGENCY_CODE_MAP = {"1": "Ordinaria", "2": "Urgente", "3": "Emergencia"}

# ---------- Helpers XML ----------
def localname(tag: str) -> str:
    return tag.split("}", 1)[1] if tag.startswith("{") else tag

def find_child_by_localname(elem, name):
    for ch in list(elem):
        if localname(ch.tag) == name:
            return ch
    return None

def find_children_by_localname(elem, name):
    return [ch for ch in list(elem) if localname(ch.tag) == name]

def find_first_text(elem, path):
    cur = elem
    for name in path:
        cur = find_child_by_localname(cur, name)
        if cur is None:
            return None
    return (cur.text or "").strip()

def find_first_attr(elem, path, attr_name):
    cur = elem
    for name in path:
        cur = find_child_by_localname(cur, name)
        if cur is None:
            return None
    return cur.attrib.get(attr_name)

def combine_date_time(date_str, time_str):
    if not date_str and not time_str:
        return None
    if date_str and time_str:
        return pd.to_datetime(f"{date_str} {time_str}", errors="coerce")
    if date_str:
        return pd.to_datetime(date_str, errors="coerce")
    return None

# ---------- Lotes ----------
def extract_lotes(cfs):
    lotes = []
    for lot in find_children_by_localname(cfs, "ProcurementProjectLot"):
        lote_id = find_first_text(lot, ["ID"])
        lote_obj = find_first_text(lot, ["ProcurementProject", "Name"])
        if lote_id or lote_obj:
            lotes.append((lote_id, lote_obj))
    return lotes

# ---------- TenderResult (con normalización de identificador) ----------
def extract_tender_results(cfs):
    results = []
    for tr in find_children_by_localname(cfs, "TenderResult"):
        id_lote = find_first_text(tr, ["AwardedTenderedProject", "ProcurementProjectLotID"])
        id_tipo = find_first_attr(tr, ["WinningParty", "PartyIdentification", "ID"], "schemeName")
        id_val  = find_first_text(tr, ["WinningParty", "PartyIdentification", "ID"])

        results.append({
            "idlote": id_lote,
            "resultado": find_first_text(tr, ["ResultCode"]),
            "fecha_acuerdo": find_first_text(tr, ["AwardDate"]),
            "ofertas_recibidas": find_first_text(tr, ["ReceivedTenderQuantity"]),
            "ofertas_pymes": find_first_text(tr, ["SMEsReceivedTenderQuantity"]),
            "adjudicatario_nombre": find_first_text(tr, ["WinningParty", "PartyName", "Name"]),
            "identificador": (id_tipo, id_val),  # (tipo, id)
            "adjudicatario_pyme": find_first_text(tr, ["SMEAwardedIndicator"]),
            "adjudicatario_ute": find_first_text(tr, ["WinningParty", "PartyLegalEntity", "CompanyTypeCode"]),
            "importe_total_sin_iva": find_first_text(tr, ["AwardedTenderedProject", "LegalMonetaryTotal", "TaxExclusiveAmount"]),
            "importe_total_con_iva": find_first_text(tr, ["AwardedTenderedProject", "LegalMonetaryTotal", "PayableAmount"]),
        })

    if not results:
        return {}

    out = {}
    fields = [k for k in results[0].keys() if k != "idlote"]

    for field in fields:
        values = []
        for r in results:
            if r["idlote"] is None:
                values.append(r[field])
            else:
                if field == "identificador":
                    tipo, val = r[field]
                    values.append((r["idlote"], tipo, val))
                else:
                    values.append((r["idlote"], r[field]))

        if len(results) == 1 and results[0]["idlote"] is None:
            out[field] = values[0]
        else:
            out[field] = values

    return out

# ---------- Extractor 23 campos + bloques ----------
def extract_preview_fields(entry):
    cfs = find_child_by_localname(entry, "ContractFolderStatus")
    if cfs is None:
        return {}

    proc_code = find_first_text(cfs, ["TenderingProcess", "ProcedureCode"])
    urg_code  = find_first_text(cfs, ["TenderingProcess", "UrgencyCode"])

    end_date = find_first_text(cfs, ["TenderingProcess", "TenderSubmissionDeadlinePeriod", "EndDate"])
    end_time = find_first_text(cfs, ["TenderingProcess", "TenderSubmissionDeadlinePeriod", "EndTime"])
    plazo_presentacion = combine_date_time(end_date, end_time)

    base = {
        "id": find_first_text(entry, ["id"]),
        "subentidad_nacional": find_first_text(cfs, ["ProcurementProject", "RealizedLocation", "CountrySubentity"]),
        "codigo_subentidad_territorial": find_first_text(cfs, ["ProcurementProject", "RealizedLocation", "CountrySubentityCode"]),
        "lotes": extract_lotes(cfs),
        "tipo_procedimiento": PROCEDURE_CODE_MAP.get(proc_code, proc_code),
        "tramitacion": URGENCY_CODE_MAP.get(urg_code, urg_code),
        "over_threshold": find_first_text(cfs, ["TenderingProcess", "OverThresholdIndicator"]),
        "organo_nombre": find_first_text(cfs, ["LocatedContractingParty", "Party", "PartyName", "Name"]),
        "organo_id": find_first_text(cfs, ["LocatedContractingParty", "Party", "PartyIdentification", "ID"]),
        "plazo_presentacion": plazo_presentacion,
    }

    tr_block = extract_tender_results(cfs)
    return {**base, **tr_block}

# ---------- Leer SOLO un atom y mostrar previews ----------
rows = []
with zipfile.ZipFile(zip_path, "r") as z:
    atoms = [n for n in z.namelist() if n.lower().endswith(".atom")]
    atom_name = atoms[0]

    with z.open(atom_name) as f:
        root = ET.parse(f).getroot()

    for entry in list(root):
        if localname(entry.tag) != "entry":
            continue
        rows.append(extract_preview_fields(entry))
        if len(rows) >= MAX_ENTRIES:
            break

df_preview = pd.DataFrame(rows)

display(df_preview.head(10))
display(df_preview[df_preview.get("resultado").notna()].head(10))
display(df_preview[df_preview["lotes"].apply(lambda x: isinstance(x, list) and len(x) > 1)].head(10))
```
